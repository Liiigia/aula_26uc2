```python
import pandas as pd  
import numpy as np  
import matplotlib.pyplot as plt  

ENDERECO_DADOS = 'https://www.ispdados.rj.gov.br/Arquivos/BaseDPEvolucaoMensalCisp.csv'


try:
    print('Obendo dados')
    # Recebendo os dados
    df_ocorrencias = pd.read_csv(ENDERECO_DADOS, sep=';', encoding='iso-8859-1')

    # Delimitando as variÃ¡veis para CISP, roubo_veiculo e recuperaÃ§Ã£o_veÃ­culo
    df_roubo_recup = df_ocorrencias[['cisp', 'roubo_veiculo', 'recuperacao_veiculos']]

    # Agrupando por CISP e somando os valores de roubo_veiculo e recuperaÃ§Ã£o_veÃ­culo
    df_total_roubo_recuperacao = df_roubo_recup.groupby('cisp', as_index=False)[['roubo_veiculo', 'recuperacao_veiculos']].sum()
    print(df_total_roubo_recuperacao)

except Exception as e:
    print('Ocorreu um erro ao obter os dados:')
```

```python
# CorrelaÃ§Ã£o
try:
    print('Calculando as correlaÃ§Ãµes')
    correlacao = np.corrcoef(df_total_roubo_recuperacao.roubo_veiculo, df_total_roubo_recuperacao.recuperacao_veiculos)[0, 1]
    print(correlacao)


except Exception as e: 
    print(f'Erro ao calcular a correlaÃ§Ã£o: {e}')
```

```python
# Ploando o grÃ¡fico de dispersÃ£o
try:
    print('Plotando o grÃ¡fico de dispersÃ£o')
    plt.scatter(df_total_roubo_recuperacao['roubo_veiculo'], df_total_roubo_recuperacao['recuperacao_veiculos'])
    plt.title(f'CorrelaÃ§Ã£o : {correlacao:.4f}')
    plt.xlabel('Roubo de VeÃ­culos')
    plt.ylabel('RecuperaÃ§Ã£o de VeÃ­culos')
    plt.show()
    

except Exception as e:
    print(f'Erro ao plotar o grÃ¡fico de dispersÃ£o: {e}')
```
