```python
# Correlação 
import pandas as pd
import numpy as np  
import matplotlib.pyplot as plt  



ENDERECO_DADOS = 'https://www.ispdados.rj.gov.br/Arquivos/BaseDPEvolucaoMensalCisp.csv'


try:
    print('Obtendo dados')
    # Recebendo os dados
    df_ocorrencias = pd.read_csv(ENDERECO_DADOS, sep=';', encoding='iso-8859-1')

    # Delimintando as variáveis para CISP, lesao_corp_dolosa e lesao_com_morte
    df_lesoes = df_ocorrencias[['cisp', 'lesao_corp_dolosa', 'lesao_corp_morte']]

    
    # Agrupando por CISP e somando as ocorrências de lesões corporal dolosa e morte
    df_tota_lesoes = df_lesoes.groupby('cisp', as_index=False)[['lesao_corp_dolosa', 'lesao_corp_morte']].sum()

    print(df_tota_lesoes.head(10))

    

except Exception as e:
    print(f'Erro ao obter dados')
```

```python
# Correlação
try:
    print('Calculando correlação')
    correlacao = np.corrcoef(df_tota_lesoes.lesao_corp_dolosa, df_tota_lesoes.lesao_corp_morte)[0, 1]
    print(correlacao)

except Exception as e:
    print(f'Erro ao calcular a correlação: {e}')
```

```python
# Plotando gráfico de dispersão
try:
    print('Plotando gráfico')
    plt.scatter(df_tota_lesoes['lesao_corp_dolosa'], df_tota_lesoes['lesao_corp_morte'])
    plt.title(f'Correlação: {correlacao:.4f}')
    plt.xlabel('Lesão Corporal Dolosa')
    plt.ylabel('Lesão Corporal com Morte')
    plt.show()

except Exception as e:
    print(f'Erro ao plotar gráfico: {e}')
```
